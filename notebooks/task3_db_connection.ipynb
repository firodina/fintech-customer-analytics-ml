```python
import sys
from pathlib import Path

# -------------------------------
# Auto-detect project root
# -------------------------------

current_dir = Path.cwd()

def find_src_folder(start_dir: Path):
    """
    Walks upward from the start directory until it finds `src`.
    Returns the src path if found, otherwise None.
    """
    for parent in [start_dir] + list(start_dir.parents):
        src_path = parent / "src"
        if src_path.exists() and src_path.is_dir():
            return src_path
    return None

src_path = find_src_folder(current_dir)

if src_path is None:
    raise FileNotFoundError("‚ùå Could not locate `src` folder. Ensure your project structure is correct.")

# Add src to Python path
if str(src_path.resolve()) not in sys.path:
    sys.path.append(str(src_path.resolve()))

print(f"‚úÖ src folder located at: {src_path.resolve()}")
print("   Added to PYTHONPATH successfully.")

# -------------------------------
# Also add task-3 for db_insert.py
# -------------------------------
task3_path = src_path / "task-3"
if task3_path.exists():
    if str(task3_path.resolve()) not in sys.path:
        sys.path.append(str(task3_path.resolve()))
        print(f"‚úÖ task-3 folder added: {task3_path.resolve()}")
else:
    print("‚ö†Ô∏è Warning: task-3 folder not found.")
```
